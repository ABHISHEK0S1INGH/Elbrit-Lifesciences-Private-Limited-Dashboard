```python
import pandas as pd
df = pd.read_csv("ELB-Sales-Data.csv")
```

```python
df
```

```python

df.drop(columns='Sr', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.astype({'Free Item Value': int, 'Rate Difference': int, 'Sales Return': int, 'Free Item': int, 'Net Rate': float, 'Quantity': int, 'Claim Offer': float, 'Breakage' : float, 'Against Expiry': float, 'Is added to primary log': int, 'Cancelled': int, 'Primary Sales': float})
```

```python
Return_Reason_new = df[['Return for Reason']]
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer( strategy='most_frequent')
df= pd.DataFrame(imputer.fit_transform(df),columns = df.columns)
```

```python
df = pd.concat([df, Return_Reason_new], axis=1)
```

```python
df.to_csv('output.csv')
```
